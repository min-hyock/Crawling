### selenium 라이브러리
- 웹 브라우저를 자동으로 제어하기 위한 라이브러리 (자동화)
- 반드시 최초 한번 설치 해주어야 함
    - 설치는 한번만
- webdriver : 브라우저 역할
- Keys : 컴퓨터용 키보드
- By : 요소 구분자 역할

```python
# pip 큰 패키지에서 설ㅇ치하겠다 라이브러리를
# ! : 강제 진행 하겠다는 뜻

%pip install selenium
```

```python
# selenium 라이브러리 로딩

# 브라우저 역할 - 컴퓨터용 브라우저
from selenium import webdriver as wb

# 키보드 역할 (Keys K대문자 !!) - 컴퓨터용 키보드
from selenium.webdriver.common.keys import Keys

# 구분자 역할 (By B대문자 !!) - 선택자를 구분
from selenium.webdriver.common.by import By
```

### 브라우저 실행
- 크롬 드라이버 설치
    - 항상 내가 사용하고 있는 크롬의 버전과 동일한 버전으로 설치해주기 !⭐
    - 크롬드라이버와 작성중인 코드(ipynb파일)를 같은 경로에 위치

```python
# wb.Chrome() : 웹 드라이버 기능 중 크롬 브라우저를 사용 -> driver 변수 담기
# wb.Chrome(설치 경로) : 경로를 생략하고 싶다면 코드와 같은 경로에 위치
# driver.get('url') : 특정 페이지를 조회

driver = wb.Chrome()
driver.get('https://www.naver.com/')

# 크롬드라이버 주의사항
# 설치유무(버전) / 설치경로 설정
```

```python
# 검색창 조회
# element, elements : 단수/복수
# find_element(By.구분자, '값')

search = driver.find_element(By.ID, 'query')
```

```python
# 검색창에 특정 값 보내기
# 요소에게 값 전송할 때 : send_keys('값')

search.send_keys('맛집')
```

### 검색 진행
- 엔터를 눌러서 검색
- 검색(돋보기) 버튼 클릭

```python
# 엔터 누르기 (값 전송과 똑같은 원리)
# Keys() 기능 활용

search.send_keys(Keys.ENTER)
```

```python
# 뒤로가기 실행하기
# 뒤로가기는 HTML 코드가 아니다
# driver.back() 기능을 직접 실행

driver.back()
```

```python
# 페이지가 변경되면(초기화) 모든 정보를 새로 받아와야 함★ 

search = driver.find_element(By.ID, 'query')
search.send_keys('카페')
```

```python
# 검색 버튼 요소 찾아오기
# -> btn 변수에 담아주기

btn = driver.find_element(By.ID, 'search_btn')
btn.click()
```

```python
# 스크롤을 페이지 최하단으로 이동
# body 변수에 담기
# body한테 end키 보내기 
body = driver.find_element(By.TAG_NAME, 'body')
body.send_keys(Keys.END)
```

```python
# 크롬 창 종료
# 주의 ! 닫기 버튼은 HTML 코드가 아니다

driver.close()
# driver.quit()

# close() : 속도가 조금 오래걸린다 ! (메모리를 완전히 종료시켜줌)
# quit() : 속도가 매우 빠름 ! (메모리가 남아있는다)
```
